## Pie Chart (3 points)

```python
# Create a pie chart from raw values

slices = [56, 66, 24, 124, 47]
labels = ["US","UK","MX","CA","JP"]
```

```python
import matplotlib.pyplot as plt

labels = 'US', 'UK', 'MX', 'CA', 'JP'
sections = [56, 66, 24, 124, 47]
colors = ['c', 'b', 'g', 'r', 'm']

plt.pie(sections, labels=labels, colors=colors,
        startangle=90,
        autopct = '%1.2f%%')

plt.axis('equal')
plt.title('Pie Chart - Countries')
plt.show()
```

```python
# exponential trendline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

num_points = 50
x = 5 + np.arange(num_points) + np.random.randn(num_points)
y = 10 + np.arange(num_points) + 5 * np.random.randn(num_points)


z = np.polyfit(x, y, 2)
p = np.poly1d(z)

plt.plot(x, p(x))
plt.scatter(x,y)
plt.title('Scatter Plot of X and Y')
```
